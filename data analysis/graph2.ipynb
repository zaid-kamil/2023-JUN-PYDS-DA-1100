# categorical graphs
- columns that have text are generally categorical
- the first step is to count the number of unique values in column
- then plot the count of each category
- following charts r a good way to visualize categorical data
    - bar chart
    - pie chart
    - donut chart
    - line chart (when too many categories)
    - area chart (when too many categories)


```python
survival_count = titanic['alive'].value_counts()
survival_count.plot(kind='pie', autopct='%.1f%%', title='Survival Rate')
```

```python
survival_count.plot(
    kind='pie', 
    autopct='%.1f%%', 
    title='Survival Rate',
    labels=['Survived', 'Not Survived'],
    colors=['#66b3ff','#ff9999'],
    labeldistance=1.1,
    wedgeprops={
        'width': 0.5,
    })
```

# seaborn
- for complex data visualization using simple code

```python
sns.countplot(data=titanic, x='alive', hue='sex')
```

```python
sns.countplot(data=titanic, x='alive', hue='pclass')
```

```python
sns.jointplot(data=titanic, x='fare', y='age')
```

```python
sns.jointplot(data=titanic, x='fare', y='age', kind='scatter', hue='survived')
```

```python
dots = sns.load_dataset("dots")
sns.relplot(
    data=dots, kind="line",
    x="time", y="firing_rate", col="align",
    hue="choice", size="coherence", style="choice",
    facet_kws=dict(sharex=False)
)
```

```python
penguins = sns.load_dataset("penguins")
```

```python
g = sns.PairGrid(penguins, hue="species", corner=True)
g.map_lower(sns.scatterplot, marker="+")
g.map_diag(sns.histplot, element="step", linewidth=0, kde=True)
g.add_legend(frameon=True)
g.legend.set_bbox_to_anchor((.61, .6))
```
